```python
import gspread
import pprint
from oauth2client.service_account import ServiceAccountCredentials
from tabulate import tabulate
import pandas as pd
```

```python
scope = ['https://spreadsheets.google.com/feeds', 'https://www.googleapis.com/auth/drive']
creds = ServiceAccountCredentials.from_json_keyfile_name('bank_secret.json', scope)
client = gspread.authorize(creds)
```

```python
def get_lowest_bank(file_name, bank_name):
    sheet = client.open(file_name.lower()).sheet1
    data = pd.DataFrame(sheet.get_all_records())

    interest_rate = data.groupby('Bank_Name')
    bank_programs = interest_rate.get_group(bank_name)
    sorted = bank_programs.sort_values(by=['interestRate'], ascending=True)
    top_lowest = sorted.iloc[0:5]
    return top_lowest
```
